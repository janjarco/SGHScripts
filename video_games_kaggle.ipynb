```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df = df.dropna(how = 'any')
df.info()
```

```python
#Relation between Platform and Genre

plat_genre = pd.crosstab(df.Platform, df.Genre)
```

```python
plat_genre_total = plat_genre.sum(axis = 1).sort_values(ascending = False)
plat_genre_total
```

```python
plt.figure(figsize = (8,6))
sns.barplot(y = plat_genre_total.index, x = plat_genre_total.values, orient = 'h')
plt.ylabel = 'PLatform'
plt.xlabel = 'The number of games'
plt.show()
```

```python
plat_genre['Total'] = plat_genre.sum(axis = 1)
plat_genre
```

```python
pop_platform = plat_genre[plat_genre.Total>1000].sort_values(by = 'Total', ascending = False)
```

```python
pop_platform
```

```python
neededdata = pop_platform.loc[:,: 'Strategy']
mini = neededdata.values.min()
maxi = neededdata.values.max()
pop_platform_final = pop_platform.append(pd.DataFrame(
    pop_platform.sum(), columns= ['Total']).T, ignore_index = False)
sns.set(font_scale=0.7)
plt.figure(figsize = (10,5))
sns.heatmap(pop_platform_final, vmin = mini, vmax = maxi, annot = True, fmt = 'd', cmap = sns.cubehelix_palette(8, start=.5, rot=-.75))
plt.xticks(rotation = 90)
plt.show()
```

```python
genre_group  = df.groupby(['Genre']).sum().loc[:, 'NA_Sales':]

genre_group['NA_Sales%'] = genre_group['NA_Sales']/genre_group['Global_Sales']
genre_group['EU_Sales%'] = genre_group['EU_Sales']/genre_group['Global_Sales']
genre_group['JP_Sales%'] = genre_group['JP_Sales']/genre_group['Global_Sales']
genre_group['Other_Sales%'] = genre_group['Other_Sales']/ genre_group['Global_Sales']
plt.figure(figsize = (7,10))
sns.set(font_scale = 0.7)
plt.subplot(211)
sns.heatmap(genre_group.loc[:, 'NA_Sales':'Other_Sales'], fmt = '.01f', annot = True,   cmap = sns.cubehelix_palette(8, start=.5, rot=-.75))
plt.title('Comparation of each region in each genre')
plt.subplot(212)
sns.heatmap(genre_group.loc[:, 'NA_Sales%':'Other_Sales%'], annot = True, vmin = 0, vmax= 1,  fmt = '.2%',  cmap = sns.cubehelix_palette(8, start=.5, rot=-.75))
plt.title('Comparation of sales in each region in each genre in percentage')
```
